### **Explanatory Notebook**

This notebook shows results of data analysis

Top features are those who contribute meaningfully to predicting the target

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("your_data.csv")
df = df.drop(columns=["Unnamed: 0"])

# Load top features
import json
with open("top_features.json", "r") as f:
    top_features = json.load(f)

print("Top features:", top_features)

# best univariate graphs
for col in top_features:
    plt.figure(figsize=(5,3))
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()      

# Best bivariate graphs
for col in top_features:
    plt.figure(figsize=(5,3))
    sns.boxplot(x="smoking", y=col, data=df)
    plt.title(f"{col} vs Smoking Status")
    plt.show()

# The heatmap
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), cmap="coolwarm", annot=True)
plt.show()
```
